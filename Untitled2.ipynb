```python
import pandas as pd
df = pd.read_excel("road_accident_data.xlsx")
df.head()
```

```python
df['Time_of_Day'] = pd.to_datetime(df['Time_of_Day'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
mode_value = df["Weather_Conditions"].mode()[0]

df["Weather_Conditions"] = df["Weather_Conditions"].fillna(mode_value)

print(df["Weather_Conditions"].isnull().sum())
```

```python
mode_value = df["Type_of_Junction"].mode()[0]

df["Type_of_Junction"] = df["Type_of_Junction"].fillna(mode_value)

print(df["Type_of_Junction"].isnull().sum())
```

```python
median_speed = df[df["Vehicle_Speed"] != 0]["Vehicle_Speed"].median()

df["Vehicle_Speed"] = df["Vehicle_Speed"].replace(0, median_speed)

print((df["Vehicle_Speed"] == 0).sum())
```

```python
median_visibility = df[df["Visibility"] < 999]["Visibility"].median()

df["Visibility"] = df["Visibility"].replace(999.9313419392096, median_visibility)

print(df["Visibility"].max())

print((df["Visibility"] == 999.9313419392096).sum())
```

```python
median_friction = df[df["Road_Surface_Friction_Coefficient"] != 0.1]["Road_Surface_Friction_Coefficient"].median()

df["Road_Surface_Friction_Coefficient"] = df["Road_Surface_Friction_Coefficient"].replace(0.1, median_friction)

print((df["Road_Surface_Friction_Coefficient"] == 0.1).sum())
```

```python
speed_mode = df["Speed_Limit"].mode()[0]

df["Speed_Limit"] = df["Speed_Limit"].fillna(speed_mode)

print((df["Speed_Limit"] == 0).sum())
```
